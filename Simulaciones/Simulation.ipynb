```python
def newpoint():
    return round(uniform(-103.39182,-103.99999),5), round(uniform(20.66682,20.999999),5)
```

```python
points = (newpoint() for x in range(100))
origen = []
for point in points:
    origen.append(point)
```

```python
points1 = (newpoint() for x in range(100))
destino = []
for point1 in points1:
    destino.append(point1)
```

```python
point1
```

```python
id1 = []
for i in range(100):
    id1.append(i)
```

```python
import pandas as pd
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib import cm
```

```python
df1 = pd.DataFrame(origen, columns=['longitud_origen','latitud_origen'])
```

```python
df2 = pd.DataFrame(destino, columns=['longitud_destino','latitud_destino'])
```

```python
df3 = pd.DataFrame(id1, columns=['node'])
```

```python
df = pd.concat([df3,df1,df2],axis=1)
```

```python
import folium
from folium.plugins import FastMarkerCluster
```

```python
folium_map= folium.Map(
    location=[18.840329972, -89.121999512],
    zoom_start=4.4,
    tiles='CartoDB dark_matter')

FastMarkerCluster(df[['latitud_origen', 'longitud_origen']].values.tolist()).add_to(folium_map)
folium.LayerControl().add_to(folium_map) 
for row in df.iterrows():
    row=row[1]
    folium.CircleMarker(location=(row["latitud_origen"],
                                  row["longitud_origen"]),
                        radius= 10,
                        color="#007849",
                        popup=row['node'],
                        fill=True).add_to(folium_map)
```
