```python
def newpoint():
    return round(uniform(-89.621667,-89.999999),6), round(uniform(20.967778,20.999999),6)
```

```python
points = (newpoint() for x in range(100))
origen = []
for point in points:
    origen.append(point)
```

```python
points1 = (newpoint() for x in range(100))
destino = []
for point1 in points1:
    destino.append(point1)
```

```python
point1
```

```python
id1 = []
for i in range(100):
    id1.append(i)
```

```python
import pandas as pd
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib import cm
```

```python
df1 = pd.DataFrame(origen, columns=['longitud_origen','latitud_origen'])
```

```python
df2 = pd.DataFrame(destino, columns=['longitud_destino','latitud_destino'])
```

```python
df3 = pd.DataFrame(id1, columns=['node'])
```

```python
df = pd.concat([df3,df1,df2],axis=1)
```

```python
def load_graph(nodes_csv_path, center, distance, network_type):
    file_h = get_graph_hash(center, distance, network_type)
    file_name = "_temp_" + file_h + ".graphml"
    cache_dir = get_cache_dir(nodes_csv_path)

    try:
        return ox.load_graphml(file_name, folder=cache_dir)
    except FileNotFoundError:
        G = ox.graph_from_point(
            center, distance=distance, simplify=False, network_type=network_type
        )
        ox.save_graphml(G, filename=file_name, folder=cache_dir)
        return G
```
